```python
import pandas as pd
```

```python
df = pd.read_csv("C:/Users/Admin/OneDrive/Documents/Book Recommendation/Audible_Catlog.csv")
df1 = pd.read_csv("C:/Users/Admin/OneDrive/Documents/Book Recommendation/Audible_Catlog_Advanced_Features.csv")
```

```python
merged_df = pd.merge(df, df1, on=["Book Name", "Author"], how="inner")
```

```python
merged_df
```

```python
merged_df["Rating"] = merged_df[["Rating_x", "Rating_y"]].max(axis=1)
merged_df.drop(columns=["Rating_x", "Rating_y"], inplace=True)
```

```python
merged_df
```

```python
merged_df["Number of Reviews"] = merged_df[["Number of Reviews_x", "Number of Reviews_y"]].max(axis=1)
merged_df["Price"] = merged_df[["Price_x", "Price_y"]].max(axis=1)
merged_df.drop(columns=["Number of Reviews_x", "Number of Reviews_y", "Price_x", "Price_y"], inplace=True)
merged_df
```

```python
merged_df.isna().sum()
```

```python
import re
import numpy as np

# Define a function to extract the rank number
def extract_rank(rank_str):
    match = re.search(r'#(\d+)', rank_str)  # Search for a number preceded by '#'
    if match:
        return int(match.group(1))  # Return the number as an integer
    else:
        return np.nan
```

```python
merged_df['Rank'] = merged_df['Ranks and Genre'].apply(extract_rank)
merged_df['Rank'].fillna(-1, inplace=True)
```

```python
merged_df['Rank'] = merged_df['Rank'].astype('int')
```

```python
def clean_genre(text):
    # Handle missing or invalid values
    if pd.isna(text) or text == '-1' or text == '':
        return None
    
    # Split only on the last occurrence of ' in ' (with spaces)
    if ' in ' in text:
        return text.rsplit(' in ', 1)[-1].strip()  # rsplit splits from the right
    return text.strip()
```

```python
merged_df['Genre'] = merged_df['Ranks and Genre'].apply(clean_genre)
```

```python
merged_df.drop(columns=['Ranks and Genre'], inplace=True)
```

```python
merged_df['Description'].fillna('No description available', inplace=True)
```

```python
merged_df
```

```python
import numpy as np

def convert_to_minutes(time_str):
    if time_str == "-1":
        return np.nan  # Treat '-1' as missing data
    hours = re.findall(r'(\d+)\s*hours?', time_str)
    minutes = re.findall(r'(\d+)\s*minutes?', time_str)
    
    total_minutes = 0
    if hours:
        total_minutes += int(hours[0]) * 60  # Convert hours to minutes
    if minutes:
        total_minutes += int(minutes[0])  # Add remaining minutes
    
    return total_minutes
```

```python
merged_df['Listening Time (minutes)'] = merged_df['Listening Time'].apply(convert_to_minutes)
```

```python
merged_df.drop(columns=['Listening Time'], inplace=True)
```

```python
merged_df['Listening Time (minutes)'].fillna(merged_df['Listening Time (minutes)'].median(), inplace=True)
merged_df
```

```python
merged_df['Number of Reviews'] = pd.to_numeric(merged_df['Number of Reviews'], errors='coerce')
merged_df['Number of Reviews'].fillna(merged_df['Number of Reviews'].median(), inplace=True)

merged_df.head()
```

```python
avg = merged_df['Rating'].mean()
merged_df['Rating'].replace(-1, avg, inplace=True)
```

```python
merged_df.drop_duplicates(inplace=True)
```
